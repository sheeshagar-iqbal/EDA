```python
from sklearn.preprocessing import StandardScaler
scaler =StandardScaler()
col =['Age', 'Cholesterol', 'Blood Pressure', 'Heart Rate', 'Exercise Hours',
       'Stress Level', 'Blood Sugar']

df[col]=scaler.fit_transform(df[col])
df.head()
```

```python
df.columns
col=['Age', 'Cholesterol', 'Blood Pressure', 'Heart Rate', 'Exercise Hours',
       'Stress Level', 'Blood Sugar', 'Heart Disease', 'Gender_Female',
       'Smoking_Current', 'Smoking_Former', 'Smoking_Never',
       'Alcohol Intake_Heavy', 'Alcohol Intake_Moderate', 'Family History_Yes',
       'Diabetes_Yes', 'Obesity_Yes', 'Exercise Induced Angina_Yes',
       'Chest Pain Type_Asymptomatic', 'Chest Pain Type_Atypical Angina',
       'Chest Pain Type_Non-anginal Pain', 'Chest Pain Type_Typical Angina']
```

```python
from scipy.stats import pearsonr
correlations ={
    feature: pearsonr(df[feature],df['Heart Disease'])[0]
 
    for feature in col
}
correlations
```

```python
corr_df = pd.DataFrame(list(correlations.items()), columns=['feature','correlations'])
corr_df
```
